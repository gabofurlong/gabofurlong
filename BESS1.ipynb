```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Parameters
initial_investment = 1000000  # Initial cost of the battery system
annual_operation_cost = 50000  # Annual maintenance and operation cost
annual_revenue = 150000  # Expected annual revenue from the system
discount_rate = 0.07  # Discount rate for NPV calculations
```

```python
years = np.arange(0, 16)
cash_flows = np.full(16, -annual_operation_cost)
cash_flows[0] = -initial_investment
cash_flows[1:] += annual_revenue
```

```python
npv = np.sum(cash_flows / (1 + discount_rate) ** years)
print(f"Net Present Value (NPV) over 15 years: ${npv:.2f}")
```

```python
plt.figure(figsize=(10, 6))
plt.bar(years, cash_flows, label='Annual Cash Flow')
plt.plot(years, np.cumsum(cash_flows), color='r', marker='o', label='Cumulative Cash Flow')
plt.axhline(0, color='black',linewidth=0.5)
plt.title('Annual and Cumulative Cash Flows')
plt.xlabel('Year')
plt.ylabel('Amount ($)')
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
