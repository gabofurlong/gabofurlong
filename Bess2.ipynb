```python
import numpy as np
import matplotlib.pyplot as plt

# Initial parameters
initial_investment = 1000000
annual_operation_cost = 50000
annual_revenue = 150000
years = 15

```

```python
num_simulations = 1000
tariff_reduction = 0.05
roi_policy_change = []

for _ in range(num_simulations):
    adjusted_revenue = annual_revenue * (1 - tariff_reduction)
    total_revenue = adjusted_revenue * years
    roi = (total_revenue - (initial_investment + annual_operation_cost * years)) / initial_investment
    roi_policy_change.append(roi)

average_roi_policy_change = np.mean(roi_policy_change)
```

```python
degradation_rates = [0.02, 0.03, 0.05]
roi_degradation = {rate: [] for rate in degradation_rates}

for rate in degradation_rates:
    for _ in range(num_simulations):
        total_revenue = 0
        for year in range(1, years + 1):
            total_revenue += annual_revenue * (1 - rate) ** year
        roi = (total_revenue - (initial_investment + annual_operation_cost * years)) / initial_investment
        roi_degradation[rate].append(roi)

average_roi_degradation = {rate: np.mean(roi_degradation[rate]) for rate in degradation_rates}
```

```python
plt.hist(roi_policy_change, bins=50, alpha=0.5, label='Policy Change')
for rate in degradation_rates:
    plt.hist(roi_degradation[rate], bins=50, alpha=0.5, label=f'Degradation {rate*100}%')
plt.xlabel('ROI')
plt.ylabel('Frequency')
plt.legend()
plt.title('ROI Distributions for Different Risk Factors')
plt.show()
```
